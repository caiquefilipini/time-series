```python
from datetime import datetime
import numpy as np
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
from matplotlib.pylab import rcParams
rcParams['figure.figsize'] = 15, 6
import warnings
warnings.filterwarnings('ignore')
```

```python
dateparse = lambda dates: datetime.strptime(dates, '%Y-%m')
data = pd.read_csv('../../data/AirPassengers.csv', parse_dates=['Month'], index_col='Month', date_parser=dateparse)
```

```python
data.head(3)
```

```python
from statsmodels.tsa.api import ExponentialSmoothing, SimpleExpSmoothing, Holt
```

```python
fit1 = ExponentialSmoothing(data, seasonal_periods=12, trend='additive', seasonal='additive').fit(use_boxcox=True)
```

```python
fit1.fittedvalues.plot(style='--', color='red');
```

```python
fit1.forecast(24).plot(style='--', marker='o', color='black', legend=True);
```
