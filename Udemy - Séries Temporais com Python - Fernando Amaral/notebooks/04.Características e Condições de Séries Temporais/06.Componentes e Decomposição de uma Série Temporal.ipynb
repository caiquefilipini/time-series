```python
import numpy as np
import pandas as pd
from datetime import datetime
import matplotlib.pylab as plt
%matplotlib inline
from pandas.plotting import register_matplotlib_converters
register_matplotlib_converters()
from matplotlib.pylab import rcParams
rcParams['figure.figsize'] = 15, 6
```

```python
dateparse = lambda dates: datetime.strptime(dates, '%Y-%m')
data = pd.read_csv('../../data/AirPassengers.csv', parse_dates=['Month'], index_col='Month', date_parser=dateparse)
```

```python
data.head()
```

```python
plt.plot(data)
plt.title('Passageiros Internacionais Transportados', fontsize=20)
plt.ylabel('Passageiros')
plt.xlabel('Tempo');
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
result = seasonal_decompose(data, model='multiplicative')
```

```python
result
```

```python
result.plot();
```

```python
result2 = seasonal_decompose(data, model='additive')
```
